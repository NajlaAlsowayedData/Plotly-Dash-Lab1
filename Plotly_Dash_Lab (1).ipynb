# Q2: Make dash that takes name input and outpt "Welcome to Data Science dashboard {name}"

```python
App = JupyterDash('Name')
App.layout = html.Div([
    html.Div(["Input Your Name: ", dcc.Input(id="Input1" , value="initial value" , type="text" )]),
    # html.H1("Welcome to Data Science: "),
    html.Br(),
    html.Div(id='my-output'),
])
@App.callback(
    Output(component_id='my-output', component_property='children'),
    Input(component_id='Input1', component_property='value')
)
def update_output_div(Input1):
    return f'Welcome to Data Science {Input1} :)'

App.run_server()
```

# Q3: Make Plotly-Dash that takes two inputs (numbers) and output 3 numbers x, y , z= x+y
- hint: use html.Div, dcc.Input(type='number')

```python
App = JupyterDash("Name")

App.layout = html.Div([
    html.H1('Addition Calculator'),
    dcc.Input(id='input1', type='number', value="number"),
    dcc.Input(id='input2', type='number', value="number"),
    html.H2(id='x'),
    html.H2(id='y'),
    html.H2(id='z')
])

@App.callback(
    [Output('x', 'children'),
     Output('y', 'children'),
     Output('z', 'children')],
    [Input('input1', 'value'),
     Input('input2', 'value')]
)
def update_output(input1, input2):
    x = input1
    y = input2
    z = x + y
    return x, y, z

App.run_server()
```

# Q6: Make Plotly chart, then add it to a plotly_dash
